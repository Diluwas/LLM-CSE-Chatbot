```python
from langchain.embeddings.openai import OpenAIEmbeddings
from openai import AzureOpenAI
from langchain.vectorstores.azuresearch import AzureSearch
import os
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
index_name = ""
azure_search_endpoint = " "
azure_search_key = ""

api_key = ""
```

```python
embeddings = OpenAIEmbeddings(api_key=api_key)
```

```python
vectore_store = AzureSearch(
    azure_search_endpoint=azure_search_endpoint,
    azure_search_key=azure_search_key,
    index_name=index_name,
    embedding_function=embeddings.embed_query
)
```

```python
def load_and_process_pdfs(pdf_floder_path):
    documents = []
    for file in os.listdir(pdf_floder_path):
        if file.endswith('.pdf'):
            pdf_path = os.path.join(pdf_floder_path, file)
            loader = PyPDFLoader(pdf_path)
            documents.extend(loader.load())
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
    splits = text_splitter.split_documents(documents)
    return splits
```

```python
pdf_folder_path = "D:\Projects\LLM-RAG-chatbot"
splits = load_and_process_pdfs(pdf_folder_path)
```

```python
vectore_store.add_documents(documents=splits)
```

```python
docs = vectore_store.similarity_search(
    query="How to start investing CSE?",
    k=20,
    search_type="hybrid",
)
print(docs[1].page_content)
```

```python
context = "\n\n".join([doc.page_content for doc in docs])
```

```python
print(context)
```

```python
azure_oai_api_key = ""
azure_oai_endpoint = ""
azure_oai_deployment_name = "gpt-4.1"

client = AzureOpenAI(
    api_key=azure_oai_api_key,
    api_version="2025-01-01-preview",
    azure_endpoint=azure_oai_endpoint,
    azure_deployment=azure_oai_deployment_name
)
response = client.chat.completions.create(
    model="gpt-4.1",
    messages=[
        {"role": "system", "content": "You are an AI assistant providing investment guidance."},
        {"role": "user", "content": f"Based on the following information:\n\n{context}\n\nWhat are tradibg sessions?"}
    ]
)

print(response.choices[0].message.content)
```
